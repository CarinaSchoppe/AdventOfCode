```python
lines = [line.strip() for line in open("adventofcode2.txt", "r")]
total_good = 0

def report_checker(elements):
    order = "ASC" if int(elements[0]) - int(elements[1]) < 0 else "DESC"
    #check if all numbers in the list are either asc or desc
    if order == "ASC":
        if all(0 > int(elements[i]) - int(elements[i + 1]) > -4 for i in range(len(elements) - 1)):
            return 1
    elif order == "DESC":
        if all(0 < int(elements[i]) - int(elements[i + 1]) < 4 for i in range(len(elements) - 1)):
            return 1

    return 0

for line in lines:
    elements = line.split(" ")
    if report_checker(elements) == 1:
        print(f"good line: {elements}")
        total_good += 1
    else:
        for i in range(len(elements)):
            elem = elements.copy()
            elem.pop(i)
            if report_checker(elem) == 1:
                total_good += 1
                print(f"removed: {elements[i]} for good in {elements}")
                break
        else:
            print(f"totally bad {elements}")



print(f"total_good: {total_good}")

```

```python
import re


def calculate(string):
    result = re.findall("mul\\((\\d{1,3}),(\\d{1,3})\\)", string)
    final = 0
    for left, right in result:
        final += int(left) * int(right)
    print(final)

with open("adventofcode3.txt", "r") as file:
    string = ""
    for line in file:
        string +=line

strings = string.split("do")
valid = [string for string in strings if not string.startswith("n't()") ]
calculate("".join(valid))
```
